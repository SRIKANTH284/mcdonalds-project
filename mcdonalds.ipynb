```python
# CATEGORICAL FEATURES
data.describe(include='object')
```

```python
# 1. COUNT PLOT
cols = 4
rows = 1
fig = plt.figure(figsize= (16,6))
all_categs = data.select_dtypes(include='object')
cat_cols = all_categs.columns[all_categs.nunique() < 10]
for i, col in enumerate(cat_cols):
    
    ax=fig.add_subplot(rows, cols, i+1)
    
    sns.countplot(x=data[col], ax=ax)
    
    plt.xticks(rotation=45, ha='right')
    
fig.tight_layout()  
plt.show()
```

```python
# 2. PIE CHART
df = data['status'].value_counts()
plt.pie(df, labels=df.index, autopct="%.0f%%");
```

```python
# 3. RUG PLOT - sns.rugplot()
sns.rugplot(x='lat', data=data, height=.03, color='darkblue')
sns.histplot(x='lon', data=data, kde=True)
```

```python
# Scatter plot
fig, ax = plt.subplots(figsize = (18,10))
ax.scatter(data['lon'], data['lat'])
 
# x-axis label
ax.set_xlabel('(lon)')
 
# y-axis label
ax.set_ylabel('(lat)')
plt.show()
```

```python
#The regplot() function takes an argument logistic , which allows you to
#specify whether you want to estimate a logistic regression model for the given data
#using true or false . This will also create a customization drawing.
sns.regplot(x="lon", y="alt", data=data)
```

```python

# Scatterplot arguments
sns.lmplot(x='lat', y='lon', data=data,
           fit_reg=False, # No regression line
           hue='alt')   # Color by evolution stage
```
